```python
#导入包
import pandas as pd
import numpy as np
#import pandas_profiling
import matplotlib.pyplot as plt
```

```python
#读取数据集
path = "C:/Github/pandas_exercises/01_Getting_&_Knowing_Your_Data/Chipotle/chipotle.xlsx"
chipo = pd.read_excel(path)
```

```python
#查看前几行数据
chipo.head()
```

```python
#查看数据类型
chipo.info()
```

```python
#发现item_price是字符型的，转成浮点型
chipo.item_price = chipo.item_price.apply(lambda s: float(s[1:]))
```

```python
#对数据集进行探索
chipo.profile_report()
```

```python
#发现有59行数据是完全重复的，做去重处理
chipo.drop_duplicates(inplace=True)
```

```python
#价格不能直接叠加，增加一个字段：total_price
chipo['total_price'] = chipo.quantity * chipo.item_price
```

```python
chipo.head()
```

```python
item_group = chipo.groupby('item_name').agg({'order_id': pd.Series.nunique, 'quantity': np.sum, 'choice_description': pd.Series.nunique, 'total_price': np.sum})
```

```python
item_group.sort_values('total_price', ascending=False)
```

```python
plt.barh(item_group.index, item_group.total_price)
```
