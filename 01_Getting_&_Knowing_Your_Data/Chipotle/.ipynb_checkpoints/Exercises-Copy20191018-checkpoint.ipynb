### Step 9. Which was the most ordered item?
销售数量(quantity）最多的item_name是什么？

```python
orderitem = chipo.groupby(chipo.item_name).sum().quantity
```

```python
orderitem.sort_values(ascending=False).head()
```

### Step 10. How many items were ordered?
有多少种不重复的item_name?

```python
len(orderitem)
```

### Step 11. What was the most ordered item in the choice_description column?
销售数量(quantity）最多的choice_description是什么？

```python
order_choice = chipo.groupby(chipo.choice_description).sum().quantity
```

```python
order_choice.sort_index(ascending=False).head()
```

### Step 12. How many items were orderd in total?
求总数量（quantity）

```python
chipo.quantity.sum()
```

### Step 13. Turn the item price into a float
item_price一列是字符串，要去掉$符号，将字符型数字转换成浮点型

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:]))
```

### Step 14. How much was the revenue for the period in the dataset?
求总订单金额

```python
totalrevenue = sum(chipo.quantity * chipo.item_price)
```

### Step 16. What is the average amount per order?
每个订单（order_id)的平均金额

```python
totalrevenue / len(chipo.groupby(chipo.order_id))
```
