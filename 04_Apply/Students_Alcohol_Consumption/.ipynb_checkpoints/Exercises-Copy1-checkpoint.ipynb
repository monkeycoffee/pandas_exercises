### Step 5. Create a lambda function that captalize strings.
意思是创建一个可以将字符串字母大写化的lambda函数

```python
cap = lambda x: x.upper() #x是单元格对象
```

```python
cap1 = lambda s: s.str.upper() #s是Series对象
```

### Step 6. Capitalize both Mjob and Fjob

```python
df[['Mjob', 'Fjob']].apply({'Mjob': cap, 'Fjob': cap})
```

```python
df[['Mjob', 'Fjob']] = df[['Mjob', 'Fjob']].apply(cap1)
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker
majority是成年的意思，即创建一个函数判断是否成年。

```python
def maj(age):
    return True if age >= 18 else False
```

```python
df['legal_drinker'] = df.age.apply(maj)
```

### Step 10. Multiply every number of the dataset by 10. 
##### I know this makes no sense, don't forget it is just an exercise
对数值型的列乘以10，其他的不变

```python
def time10(s):
    return s * 10 if s.dtype == np.int64 else s
```

```python
df.apply(time10)
```
